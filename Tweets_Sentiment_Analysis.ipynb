```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
count = 200000
tweets = []

keywords = ["(Driverless OR Without A Driver OR No Driver OR No Human OR No Operator OR Unmanned OR Uncrewed) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) \
             lang:en",
           "(Autonomous) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en",
           "(Self-Driving OR Self Drive) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en",
	         "(Automated) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en",
	         "(Computer Driven OR Computer Controlled OR Robot) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en"]

for keyword in keywords:
  for i, tweet in enumerate(sntwitter.TwitterSearchScraper(keyword).get_items()):
    if len(tweets) == count:
      break
    else:
      tweets.append([tweet.id,
                   tweet.date,
                   tweet.user.username,
                   tweet.likeCount,
                   tweet.retweetCount,
                   tweet.content,
                   keyword])

df = pd.DataFrame(tweets, columns=['TweetID', 'Tweet_Date', 'Tweet_Author', 'Tweet_Likes', 'Retweets', 'Tweet_Text', 'keyword_query'])
print(df)
```

```python
import re

def vader_preprocess(data):
  data = re.sub(r"@[A-Za-z0-9_]+", "", data)
  data = re.sub(r"http\S+|www\. \S+", "", data)
  data = re.sub(r"#", "", data)
  return data

df['Vader_CleanTweet'] = df['Tweet_Text'].apply(lambda x : vader_preprocess(x))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

VADER_model = SentimentIntensityAnalyzer()

df['Vader_Scores'] = df['Vader_CleanTweet'].apply(lambda x: VADER_model.polarity_scores(x))
```

```python
def Sentiment(text):

    VADER_model = SentimentIntensityAnalyzer()
    score_dict = VADER_model.polarity_scores(text)
    if  float(score_dict["compound"]) >= -1.0 and float(score_dict["compound"]) <= -0.25:
        Sen_label = 'Negative'
    elif float(score_dict["compound"]) >= -0.25 and float(score_dict["compound"]) <= 0.25:
        Sen_label = 'Neutral'
    else:
        Sen_label = 'Positive'
    result = Sen_label
    return result

df["Vader_sentiment"] = df["Vader_CleanTweet"].apply(lambda x : Sentiment(x))
```

```python
import torch
import torch.nn as nn
from transformers import AutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
from scipy.special import softmax
```

```python
transformer_model = f"cardiffnlp/twitter-roberta-base-sentiment-latest"
tok = AutoTokenizer.from_pretrained(transformer_model)
config = AutoConfig.from_pretrained(transformer_model)
```

```python
roberta_model = AutoModelForSequenceClassification.from_pretrained(transformer_model)
```

```python
def Pre_process(tweet):
  words_list = []
  for word in tweet.split(' '):
    if word.startswith('@') and len(word) > 1:
        word = '@user'
    elif word.startswith('http'):
        word = "http"
    words_list.append(word)

  final_tweet = " ".join(words_list)
  return final_tweet

df['Final_Tweet'] = df['Tweet_Text'].apply(lambda x : Pre_process(x))
```

```python
def roberta_sentiment(text):

    vectors = tok(text, return_tensors='pt')
    output = roberta_model(**vectors)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict

df["roberta_scores"] = df["Tweet_Text"].apply(lambda x : roberta_sentiment(x))
```
