```python
count = 200000
tweets = []

keywords = ["(Driverless OR Without A Driver OR No Driver OR No Human OR No Operator OR Unmanned OR Uncrewed) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) \
             lang:en until:2023-02-23 since:2023-02-15",
           "(Autonomous) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en until:2023-02-23 since:2023-02-15",
           "(Self-Driving OR Self Drive) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en until:2023-02-23 since:2023-02-15",
	         "(Automated) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en until:2023-02-23 since:2023-02-15",
	         "(Computer Driven OR Computer Controlled OR Robot) AND (Car OR Suv OR Vehicle OR Automobile OR Bus OR Truck OR Van OR Motor Vehicle OR Vehicle Technology OR Transport) lang:en until:2023-02-23 since:2023-02-15"
]
for keyword in keywords:
  for i, tweet in enumerate(sntwitter.TwitterSearchScraper(keyword).get_items()):
    if len(tweets) == count:
      break
    else:
      tweets.append([tweet.id,
                   tweet.date, 
                   tweet.user.username,
                   tweet.likeCount,
                   tweet.retweetCount, 
                   tweet.content])
        
df = pd.DataFrame(tweets, columns=['TweetID', 'Tweet_Date', 'Tweet_Author', 'Tweet_Likes', 'Retweets', 'Tweet_Text'])
print(df)
```

```python
import re

def vader_preprocess(data):
  #userhandles
  data = re.sub(r"@[A-Za-z0-9_]+", "", data)
  #URLs
  data = re.sub(r"http\S+|www\. \S+", "", data)
  #hashtags
  data = re.sub(r"#", "", data)

  return data 

keyword_df['Vader_CleanTweet'] = keyword_df['Tweet_Text'].apply(lambda x : vader_preprocess(x))
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

VADER_model = SentimentIntensityAnalyzer()

keyword_df['Vader_Scores'] = keyword_df['Vader_CleanTweet'].apply(lambda x: VADER_model.polarity_scores(x))
```

```python
def Sentiment(text):
    VADER_model = SentimentIntensityAnalyzer()
    score_dict = VADER_model.polarity_scores(text)

    if  float(score_dict["compound"]) >= -1.0 and float(score_dict["compound"]) <= -0.25:
        Sen_label = 'Negative'
    elif float(score_dict["compound"]) >= -0.25 and float(score_dict["compound"]) <= 0.25:
        Sen_label = 'Neutral'
    else:
        Sen_label = 'Positive'

    result = Sen_label
    return result

keyword_df["Vader_sentiment"] = keyword_df["Vader_CleanTweet"].apply(lambda x : Sentiment(x))
```
