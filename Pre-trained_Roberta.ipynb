```python
def Pre_process(tweet):
  words_list = []

  for word in tweet.split(' '):

    if word.startswith('@') and len(word) > 1:
        word = '@user'
    elif word.startswith('http'):
        word = "http"
    words_list.append(word)

  final_tweet = " ".join(words_list)
  return final_tweet
```

```python
df['Final_Tweet'] = df['Tweet'].apply(lambda x : Pre_process(x))
```

```python
import re

def vader_preprocess(data):
  #userhandles
  data = re.sub(r"@[A-Za-z0-9_]+", "", data)
  #URLs
  data = re.sub(r"http\S+|www\. \S+", "", data)
  #hashtags
  data = re.sub(r"#", "", data)

  return data

df['Vader_FinalTweet'] = df['Tweet'].apply(lambda x : vader_preprocess(x))
```

```python
#extracting sentiment out of text for VADER without the cut method
def Sentiment(text):
    VADER_model = SentimentIntensityAnalyzer()
    score_dict = VADER_model.polarity_scores(text)

    if  score_dict["compound"] >= -1.0 && score_dict["compound"] <= -0.25:
        Sen_label = 'Negative'
    elif score_dict["compound"] >= -0.25 && score_dict["compound"] <= 0.25:
        Sen_label = 'Neutral'
    else:
        Sen_label = 'Positive'

        result = Sen_label
    return result

df["vader_sentiment"] = df["Vader_FinalTweet"].apply(lambda x : Sentiment(x))
```

```python
tok = AutoTokenizer.from_pretrained(transformer_model)
config = AutoConfig.from_pretrained(transformer_model)
```

```python
roberta_model = AutoModelForSequenceClassification.from_pretrained(transformer_model)
```

```python
def roberta_sentiment(text):

    vectors = tok(text, return_tensors='pt')
    output = roberta_model(**vectors)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
def roberta_Sentiment(polarity):

  if polarity <= -1.0 & polarity >= -0.25:
    Sen_label = 'Negative'
  elif polarity <= 0.25 & polarity >= -0.25:
    Sen_label = 'Neutral'
  else:
    Sen_label = 'Positive'

  return Sen_label

scores_df["roberta_sentiment"] = scores_df["roberta_polarity"].apply(lambda x : roberta_Sentiment(x))
```
